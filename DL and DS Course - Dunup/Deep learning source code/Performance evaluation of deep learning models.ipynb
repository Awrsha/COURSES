# روش اتوماتیک

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# وارد کردن کتابخانه
import pandas as pd
import numpy as np

# وارد کردن دیتاست دیابت
dataset = pd.read_excel("C:\\Users\\ShahinN\\Desktop\\pima-indians-diabetes.xlsx", header=None)
dataset.head()                        
```

```python
# بخش بندی دیتاست
X = dataset.values[:,0:8]
Y = dataset.values[:,8]
```

```python
# تعریف مدل
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer= 'uniform' , activation= 'relu' ))
model.add(Dense(8, kernel_initializer= 'uniform' , activation= 'relu' ))
model.add(Dense(1, kernel_initializer= 'uniform' , activation= 'sigmoid' ))
```

```python
# Compile model
model.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ])
```

```python
# Fit the model
model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10)
```

# روش دستی

```python
# بخش بندی دیتا : 66% برای آموزش و 33% برای تست

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
```

```python
# تعریف مدل
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer= 'uniform' , activation= 'relu' ))
model.add(Dense(8, kernel_initializer= 'uniform' , activation= 'relu' ))
model.add(Dense(1, kernel_initializer= 'uniform' , activation= 'sigmoid' ))
```

```python
# Compile model
model.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ])
```

```python
# Fit the model
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=10)
```

# 3.ارزیابی یک مدل یادگیری عمیق با k-fold cross validation.

```python
from sklearn.model_selection import StratifiedKFold
```

```python
#  10-fold cross validation test تعریف
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
cvscores = []
for train, test in kfold.split(X, Y):
    
    # تعریف مدل
    model = Sequential()
    model.add(Dense(12, input_dim=8, kernel_initializer= 'uniform' , activation= 'relu' ))
    model.add(Dense(8, kernel_initializer= 'uniform' , activation= 'relu' ))
    model.add(Dense(1, kernel_initializer= 'uniform' , activation= 'sigmoid' ))
    
    # Compile model
    model.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ])
    
    # Fit the model
    model.fit(X[train], Y[train],  epochs=150, batch_size=10, verbose=0)
    
    # ارزیابی مدل
    scores = model.evaluate(X[test], Y[test], verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    
    cvscores.append(scores[1] * 100)
    
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```
